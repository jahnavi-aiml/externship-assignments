```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data["Diabetes"]=le.fit_transform(data["Diabetes"])
```

```python
data.head()
```

```python
x=data.iloc[:,0:8].values
y=data.iloc[:,8].values
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10000)
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression()
logreg.fit(x_train,y_train)
```

```python
ypred=logreg.predict(x_test)
```

```python
ypred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(ypred,y_test)
```

```python
accuracy
```

```python
a=logreg.predict(sc.transform([[1,163,86,36,0,23.5,0.63,32]]))
```

```python
a
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,ypred)
```

```python
cm
```

```python
import sklearn.metrics as metrics
lfpr,ltpr,threshold=metrics.roc_curve(y_test,ypred)
roc_auc=metrics.auc(lfpr,ltpr)
```

```python
roc_auc
```

```python
import matplotlib.pyplot as plt
plt.plot(lfpr,ltpr,"b",label="auc=%0.2f"%roc_auc)
plt.legend()
plt.show()
```
