```python
x=data.iloc[:,0:4].values
y=data.iloc[:,4].values
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.preprocessing import OneHotEncoder#creates columns based on categories of textual data
from sklearn.compose import ColumnTransformer
ct=ColumnTransformer([("oh",OneHotEncoder(),[3])],remainder="passthrough")#add columns to first
x=ct.fit_transform(x)
```

```python
x
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
mlr=LinearRegression()
mlr.fit(x_train,y_train)
```

```python
ypred=mlr.predict(x_test)
```

```python
ypred
```

```python
y_test
```

```python
x_test
```

```python
from sklearn.metrics import r2_score
accuracy=r2_score(ypred,y_test)
```
