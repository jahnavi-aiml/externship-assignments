```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
sns.scatterplot(x="total_bill",y="tip",data=tips)#using seaborn
```

```python
plt.scatter(tips["total_bill"],tips["tip"])#using matplot
```

```python
sns.lineplot(x="size",y="tip",data=tips)#for comparing two numerical values
```

```python
sns.distplot(tips["tip"])#showws the max and min if values and area under the graph is n1
```

```python
sns.barplot(x="time",y="tip",data=tips)#at which time tip is high
```

```python
sns.boxplot(x="size",y="tip",data=tips)#gives the min and max tips given by grp of people
```

```python
sns.catplot(x="smoker",y="tip",data=tips)
```

```python
sns.pairplot(tips)
```

```python
sns.heatmap(tips.corr(),annot=True)
```
