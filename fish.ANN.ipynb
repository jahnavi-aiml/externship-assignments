```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
cat=Sequential()
```

```python
cat.add(Dense(6,kernel_initializer="random_uniform",activation="relu"))
```

```python
cat.add(Dense(12,kernel_initializer="random_uniform",activation="relu"))
```

```python
cat.add(Dense(7,kernel_initializer="random_uniform",activation="softmax"))
```

```python
cat.compile("adam","categorical_crossentropy",metrics=["accuracy"])
```

```python
cat.fit(x_train,y_train,batch_size=32,epochs=700)
```

```python
ypred=cat.predict(x_test)
```
